```python
import pandas as pd
```

```python
df = pd.read_parquet("./data/tasks/face_task.parquet")
```

```python
df_pos = df[df["positive"] == True]
df_neg = df[df["positive"] == False]
```

```python
import numpy as np
np.random.seed(0)

def sample(df, n):
	return df.loc[np.random.choice(df.index.values, n, replace=False)]

N = 1_000
df_pos_sampled = sample(df_pos, N)
df_neg_sampled = sample(df_neg, N)
```

```python
df_neg_sampled
```

```python
pd.concat((df_pos_sampled, df_neg_sampled)).to_parquet("./data/tasks/face_task2k.parquet", index=False)
```
